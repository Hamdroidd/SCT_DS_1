```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and preview
df = pd.read_csv("0a1206eb-5b37-4b47-bd17-843f02c82e7d_Data.csv")
df.columns = df.columns.str.strip()  # Remove any leading/trailing spaces
print("Columns:", df.columns.tolist())  # Diagnose column names

# Try auto-matching the closest column names (you can adjust after seeing output)
country_col = [col for col in df.columns if 'country' in col.lower()][0]
series_col = [col for col in df.columns if 'series' in col.lower()][0]

# Filter for India and homicide series
india = df[
    (df[country_col] == 'India') &
    (df[series_col].str.contains('Intentional homicides', case=False, na=False))
]

# Extract and reshape year data
year_cols = [col for col in df.columns if '[YR' in col]
india_data = india[year_cols].transpose()
india_data.columns = ['Homicide Rate']
india_data.index = india_data.index.str.extract(r'(\d{4})')[0]
india_data = india_data.apply(pd.to_numeric, errors='coerce').dropna()

# Plot
plt.figure(figsize=(10, 5))
plt.plot(india_data.index, india_data['Homicide Rate'], marker='o')
plt.title("Intentional Homicides (Male) in India")
plt.xlabel("Year")
plt.ylabel("Homicide Rate (per 100,000 male)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
